```python
import os
from google.colab import userdata
os.environ["OPENAI_API_KEY"] = userdata.get('OPENAI_API_KEY')
```

```python
from langchain.schema import Document
from ragas.llms import LangchainLLMWrapper
from ragas.embeddings import LangchainEmbeddingsWrapper
from langchain_openai import ChatOpenAI
from langchain_openai import OpenAIEmbeddings
from ragas.testset import TestsetGenerator

def generate_Q_A_testset(docs,testset_size):
  generator_llm = LangchainLLMWrapper(ChatOpenAI(model="gpt-4o-mini"))
  generator_embeddings = LangchainEmbeddingsWrapper(OpenAIEmbeddings())

  generator = TestsetGenerator(llm=generator_llm, embedding_model=generator_embeddings)
  dataset = generator.generate_with_langchain_docs(docs, testset_size=testset_size)

  return dataset
```

```python
file_path = "/content/deepseekv3-explained.md"

with open(file_path, "r", encoding="utf-8", errors="ignore") as f:
    text = f.read()

docs = [Document(page_content=text, metadata={"source": file_path})]

dataset = generate_Q_A_testset(docs,100)
df = dataset.to_pandas()
df.to_csv('deepseekv3-explained_Q&A.csv', index=False)
```
